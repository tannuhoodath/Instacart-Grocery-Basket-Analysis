## Q. Create a loyalty flag for existing customers using the transform() and loc() functions.

```python
# creating a new column 'max_order' which will display the maximum number of orders placed by each user. 
df_ords_prods_merge['max_order'] = df_ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
# creating a new column 'loyalty_flag' to flag customer loyalty by using the values in 'max_order'
df_ords_prods_merge.loc[df_ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal Customer'
```

```python
df_ords_prods_merge.loc[(df_ords_prods_merge['max_order'] <= 40) & (df_ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular Customer'
```

```python
df_ords_prods_merge.loc[df_ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New Customer'
```

## Q. The team now wants to target different types of spenders in their marketing campaigns. This can be achieved by looking at the prices of the items people are buying. Create a spending flag for each user based on the average price across all their orders using the following criteria:
## If the mean of the prices of products purchased by a user is lower than 10, then flag them as a “Low spender.”
## If the mean of the prices of products purchased by a user is higher than or equal to 10, then flag them as a “High spender.”

```python
# Finding the mean for all users and creating a column to display the mean
df_ords_prods_merge['average_spent'] = df_ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
# checking column creation
df_ords_prods_merge[['prices', 'average_spent']].head(10)
```

```python
# creating a new column 'spending_flag' by using the results of column 'average_spent'
df_ords_prods_merge.loc[df_ords_prods_merge['average_spent'] < 10, 'spending_flag'] = 'Low Spender'
df_ords_prods_merge.loc[df_ords_prods_merge['average_spent'] >= 10, 'spending_flag'] = 'High Spender'
```

```python
# checking column creation
df_ords_prods_merge.columns
```

```python
df_ords_prods_merge[['user_id', 'average_spent', 'spending_flag']].head(15)
```

```python
# checking the frequency for High and Low Spenders
df_ords_prods_merge['spending_flag'].value_counts(dropna = False)
```

## Q. In order to send relevant notifications to users within the app (for instance, asking users if they want to buy the same item again), the Instacart team wants you to determine frequent versus non-frequent customers. Create an order frequency flag that marks the regularity of a user’s ordering behavior according to the median in the “days_since_prior_order” column. The criteria for the flag should be as follows:
## If the median of “days_since_prior_order” is higher than 20, then the customer should be labeled a “Non-frequent customer.”
## If the median is higher than 10 and lower than or equal to 20, then the customer should be labeled a “Regular customer.”
## If the median is lower than or equal to 10, then the customer should be labeled a “Frequent customer.”

```python
# creating a new column that records the median of 'days_since_prior_order'
df_ords_prods_merge['order_frequency'] = df_ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
# checking column creation
df_ords_prods_merge.columns
```

```python
df_ords_prods_merge[['order_frequency']].head(10)
```

```python
# creating a new column 'order_frequency_flag' by using the values from 'order_frequency'
df_ords_prods_merge.loc[df_ords_prods_merge['order_frequency'] > 20, 'order_frequency_flag'] = 'Non-frequent Customer'
df_ords_prods_merge.loc[(df_ords_prods_merge['order_frequency'] > 10) & (df_ords_prods_merge['order_frequency'] <= 20), 'order_frequency_flag'] = 'Regular Customer'
df_ords_prods_merge.loc[df_ords_prods_merge['order_frequency'] <= 10, 'order_frequency_flag'] = 'Frequent Customer'
```

```python
# checking column creation
df_ords_prods_merge.columns
```

```python
df_ords_prods_merge['order_frequency_flag'].value_counts(dropna = False)
```

```python
df_ords_prods_merge[['user_id', 'loyalty_flag', 'spending_flag', 'order_frequency_flag']].head(20)
```
