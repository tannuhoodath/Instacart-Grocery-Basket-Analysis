# If statement with user defined functions

```python
# Creating a new function based on the price range of the products.
def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
# Creating a new variable in the dataframe using the 'price_label' function we created
df['price_range'] = df.apply(price_label, axis = 1)
```

```python
df['price_range'].value_counts(dropna = False)
```

### Q: Suppose your clients have changed their minds about the labels you created in your “busiest_day” column. Now, they want “Busiest day” to become “Busiest days” (plural). This label should correspond with the two busiest days of the week as opposed to the single busiest day. At the same time, they’d also like to know the two slowest days. Create a new column for this using a suitable method.

```python
# writing a for-loop statement to fulfil the above criteria
result1 = []

for value in df_ords_prods_merge['orders_day_of_the_week']:
    if value in [0,1]:
        result1.append('Busiest Days')
    elif value in [4,3]:
        result1.append('Slowest Days')
    else:
        result1.append('Regularly Busy Days')
```

```python
# creating a new column called 'busiest_days'
df_ords_prods_merge['busiest_days'] = result1
```

### Q: When too many users make Instacart orders at the same time, the app freezes. The senior technical officer at Instacart wants you to identify the busiest hours of the day. Rather than by hour, they want periods of time labeled “Most orders,” “Average orders,” and “Fewest orders.” Create a new column containing these labels called “busiest_period_of_day.”

```python
# checking the frequency for 'order_hour_of_day' column
df_ords_prods_merge['order_hour_of_day'].value_counts(dropna = False)
```

#### To create 3 categories: 'Most orders', 'Average orders' and 'Fewest orders', let's split the 24 hours into 3 which means 8 hours in each category. We can assign hours to each category based on the frequency of orders for each hour. 

```python
# writing a for-loop statement for displaying busiest hours
result2 = []

for value in df_ords_prods_merge['order_hour_of_day']:
    if value in (10,11,14,15,13,12,16,9):
        result2.append('Most Orders')
    elif value in (3,4,2,5,1,0,6,23):
        result2.append('Fewest Orders')
    else:
        result2.append('Average Orders')
```

```python
# assigning result2 to a new column called 'busies_hours'
df_ords_prods_merge['busiest_hours'] = result2
```
