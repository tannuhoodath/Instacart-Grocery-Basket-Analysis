```python
# Dropping '_merge' column from 'df_ords_prods' dataframe because it will be confusing when we merge this dataframe with another in the next steps
df_ords_prods = df_ords_prods.drop(columns = ['_merge'])
```

# Combining the datasets

### We should use the merge function to combine the datasets since the information in both datasets is different, and there is a key column (product_id) in both datasets.

```python
# merging the df_ords_prods and df_prods dataframes
df_combined = df_prods.merge(df_ords_prods, on = ['product_id'], indicator = True)
```

```python
df_combined['_merge'].value_counts()
```

#### We have used an inner join, which is the default. Hence, only the rows which are present in both dataframes are merged together.
